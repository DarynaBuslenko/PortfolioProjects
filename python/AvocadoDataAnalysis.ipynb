```python
df.columns 
```

```python
# We have a Total Volume column, which should be equal to the sum of all avocados sold (bags and individual). 
#Let's check it.
df[['Small Size', 'Large Size', 'XLarge Size', 'Total Bags']].sum(axis=1)
```

```python
df['Total Volume'].equals(df[['Small Size', 'Large Size', 'XLarge Size', 'Total Bags']].sum(axis=1))
```

```python
# We have "False" above. The condition is not fulfilled everywhere. 
#Let's see how many data are not executed. First, let's write the calculation in a separate column.
```

```python
df['Check Sum'] = df[['Small Size', 'Large Size', 'XLarge Size', 'Total Bags']].sum(axis=1)

# We create new column Check Sum and write the data in this column. Our sum is saved in this new column
```

```python
# Now we need to compare our values in the Total Volume and Check Sum columns. 
# Let's do it with a special method from the numpy library.
```

```python
import numpy as np
```

```python
df['Check Passed'] = np.isclose(df['Total Volume'], df['Check Sum'])
```

```python
np.isclose(12.5, 12.500000001), 12.5 == 12.500000001

# "isclose" is not so strict equal as ==. It helps us to compare float numbers
```

```python
df['Check Passed'].value_counts()
```

```python
# Based on above results we have only 176 False (0,97%). 
# It's very small amount, we can remove it if we need this for future analysis.
```

```python
# Let's look at the data where we failed the check
df[df['Check Passed']==False]

# We compare 2 columns: Check Sum and Total Volum. The values are very close to each other
```

```python
# Also we can calculate the difference between Check Sum and Total Volum columns

filtered = df[df['Check Passed']==False] 
(
    (filtered['Total Volume'] - filtered['Check Sum'])/filtered['Check Sum']
).hist();
```

```python
# Looking at the histogram we can see that the difference is small. 
# So instead of the previous column, we will write our calculated data.
```

```python
df.rename(columns = {'Total Volume':'Total Volume Errors', 'Check Sum': 'Total Volume'}, inplace=True)
```

```python
# Let's visualize and compare the dynamics of avocado prices in general, as well as by type.
```

```python
# Let's convert the "Date" column into a datetime object and sort the DataFrame by date

df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df = df.sort_values(by='Date')
```

```python
# Import libraries for visualization

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Here we group our dataset by data and take avg price of Average Price

plt.figure(figsize=(12, 6))
plt.plot(df.groupby('Date').AveragePrice.mean());
```

```python
# In the graph above we see an increase in 2016 from May to September.
# In early 2017, prices began to rise and reached their peak in September and October of that year.
```

```python
# Let's look at the price distribution of our avocado

plt.figure(figsize=(12,5))
plt.title("Distribution Price")
ax = sns.distplot(df["AveragePrice"], color = 'r')
```

```python
# Here's another graph that displays the average price of items categorized by date.
# Central line - shows us the avg price of Average Price
# There is an aura around this line that shows the confidence interval.
# This helps us to understand in what limits the price can be

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='AveragePrice', data=df, alpha=0.3)
plt.title('Price Trend');
```

```python
# Now let's see the dynamics by types of avocados

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='AveragePrice', hue='type', data=df, alpha=0.3)
plt.title('Price Trend by Type');
```

```python
# Conventional: At the end of 2017 we can see a large drop in prices.
# Organic: Same happens with organic avocados, at the end of 2017 there is a big drop.
```

```python
# Let's check the price dynamics for avocados by type for a Chicago region:

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='AveragePrice', hue='type', data=df[df['region']=='Chicago'])
plt.title('Price Trend by Type');
```

```python
# Organic avocado is more expensive than conventional.
# Between May 2015 and May 2016, there was a significant difference in the prices of two varieties of avocados.
# Since 2017, the price gap has narrowed.
# Approx in October conventional avocado even was more expensive than organic.
```
